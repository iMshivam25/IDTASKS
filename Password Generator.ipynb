```python
import tkinter as tk
from tkinter import messagebox
import string
import random
```

```python
def generate_password():
    password_length = length.get()
    if not password_length:
        messagebox.showerror("Error length","Please enter Password length")
    if not password_length.isdigit():
        messagebox.showerror("Error valid","Enter a valid length (int)")

    password_length = int(password_length)
    
    if password_length < 8:
        response = messagebox.askquestion("Warning","Password length too short. Run anyway?")
        if response == 'no':
            return
    if password_length >100:
        messagebox.showerror("Error","Password length must below 8-100")
        return
              
    include_uppercase = uppercase_var.get()
    include_special_chars = special_chars_var.get()
    include_numbers = numbers_var.get()
    must_include_chars = must_include.get()
    
    characters = string.ascii_lowercase
    if include_uppercase:
        characters += string.ascii_uppercase
    if include_special_chars:
        characters += string.punctuation
    if include_numbers:
        characters += string.digits       
    if must_include_chars:
        for char in must_include_chars:
            if char not in characters:
                characters += char
    
    if password_length < len(must_include_chars):
        messagebox.showerror("Error", "Password length is too short for the required characters.")
        return
    
    password = must_include_chars
    remaining_length = password_length - len(password)
    
    password += ''.join(random.choice(characters) for _ in range(remaining_length))
    password = ''.join(random.sample(password, len(password)))
    password_entry.delete(0, tk.END)
    password_entry.insert(0, password)
```

```python
# Create the main window
root = tk.Tk()
root.title("Password Generator [InternDev]")
root.geometry("400x500")

padx_value = 15
pady_value = 5

#Length
t_label = tk.Label(root, text="Password Length:", font=("Arial", 18))
t_label.grid(row=0, column=0, sticky="w", padx=padx_value, pady=(25, pady_value))
length = tk.Entry(root)
length.grid(row=0, column=1, padx=padx_value, pady=(25,pady_value))

#Uppercase
uppercase_var = tk.BooleanVar()
tk.Checkbutton(root, text="Include Uppercase Letters", variable=uppercase_var).grid(row=1, column=0, columnspan=2, sticky="w", padx=padx_value, pady=(20,pady_value))

#SpecialCharsNs
special_chars_var = tk.BooleanVar()
tk.Checkbutton(root, text="Include Special Characters", variable=special_chars_var).grid(row=2, column=0, columnspan=2, sticky="w", padx=padx_value, pady=pady_value)

#Numbers
numbers_var = tk.BooleanVar()
tk.Checkbutton(root, text="Include Numbers", variable=numbers_var).grid(row=3, column=0, columnspan=2, sticky="w", padx=padx_value, pady=pady_value)

#MustInclude
tk.Label(root, text="Must Include Characters:").grid(row=4, column=0, sticky="w", padx=padx_value, pady=(20,pady_value))
must_include = tk.Entry(root)
must_include.grid(row=4, column=1, padx=padx_value, pady=(20,pady_value))

# Generate Password Button
generate_button = tk.Button(root, text="Generate Password", command=generate_password)
generate_button.grid(row=5, column=0, columnspan=2, padx=padx_value, pady=(20,pady_value))

# Display Generated Password
pswd = tk.Label(root, text="Generated Password:",font=("Arial",16))
pswd.grid(row=6, column=0,columnspan=2, sticky="w", padx=padx_value+50, pady=(20,pady_value))
password_entry = tk.Entry(root,font=("Arial",18))
password_entry.grid(row=7, column=0,columnspan=2, padx=padx_value, pady=(10,pady_value),ipady=10,ipadx=45)

root.mainloop()
```
